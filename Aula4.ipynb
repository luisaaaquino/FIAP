```python
# 1. Carga de bibliotecas
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# 2. Carga de dados
df = pd.read_csv('https://raw.githubusercontent.com/diogenesjusto/FIAP/master/SHIFT/Data/pib.csv')
```

```python
# 3. Separação de treino e teste
treino = df.iloc[0:132]
teste = df.iloc[132:138]
```

```python
4. #Regressão linear
Y_treino = treino.iloc[:, [2]]

#X =['BRP'] #Regressão linear simples
#X =['BRL','BRP'] #Regressão linear Múltipla ou Multivariada
#X =['PIBi1','PIBi2', 'PIBi11'] #Modelo Autoregressivo de ordem 2 com ajuste de sazonalidade 12
X = ['BRL','BRP','D1','D5','D11']
X_treino= treino.loc[:, X]


modelo = LinearRegression()
modelo.fit(X=X_treino, y=Y_treino)

```

```python
#Visualizar estatísticas da regressão - coeficientes
print("Coeficientes")
print(modelo.intercept_)
print(modelo.coef_)
print("R2")
print(modelo.score(X_treino, Y_treino))
```

```python
teste

#Predict (previsão) para 1 mês
#prevJan15 = a * x + b
# = 0,63*177,84+53,717

X_teste = teste.loc[:, X]

Y_prev = modelo.predict(X_teste)
Y_real = teste.loc[:,['PIB']]
Erro = Y_prev - Y_real
abs(Erro).mean() #Abs = valor absoluto ou módulo, ou seja, sem sinal. Ou ainda, raiz quadrada do quadrado | Mean = média. Juntando tudo: RMSE
```
